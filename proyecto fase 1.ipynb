#Proyecto de análisis del precio del petróleo y su correlación con diferentes evento climáticos, económicos y geopolíticos que se desarrollaron en el mundo de 1990 a 2022



## Importación de librerias

```python
import pandas as pd
import datetime
```

El archivo cop.csv contiene los cambios del precio del petroleo a través de los años. Nosotros modificamos los datos encontrados para solo mantener los datos de 1990 hasta 2022. Además, modificamos el dato date para que sea de tipo datetime y lo seteamos como index.

```python
df_crude_oil = pd.read_csv('cop.csv') #leemos el archivo

df_crude_oil['date'] = pd.to_datetime(df_crude_oil['date']) #convertimos la columna date a datetime

df_crude_oil.set_index('date', inplace = True) #seteamos la columna como index

fecha_inicio = '1990-01-01 00:00:00'
fecha_fin = '2022-11-30 00:00:00'

df_crude_oil = df_crude_oil[fecha_inicio:fecha_fin] #cortamos el df en las fechas requeridas
df_crude_oil
```

El archivo wid.csv contiene datos de los eventos globales más importantes del mundo. El dataset contiene diferentes detalles, por tanto, eliminamos columnas que considerabamos no importantes para el proyecto, además, cortamos el dataset en el rango de 1990 a 2022.

```python
df_wid = pd.read_csv('wid.csv') #lectura del archivo

df_wid.drop(index = [0,1], inplace = True) #eliminamos los primeros dos elementos pq eran BC e iban a dar problema

df_wid['Year'] = df_wid['Year'].astype(str) #pasamos los años a string

df_wid = df_wid[('1990' <= df_wid['Year']) & (df_wid['Year'] <= '2024')] #cortamos al rango de años necesitado

df_wid['Month']=df_wid['Month'].replace('Unknown', 'January') #Se tenian varios unknown en los meses, reemplazamos por enero
df_wid.drop(columns=['Date'], inplace=True) #eliminamos la columna Date, ya que no necesitamos el dia exacto

df_wid['date'] = df_wid['Year'] + '-' + df_wid['Month'] + '-01' #armamos las fechas para convertirlo en datetime

df_wid['date'] = pd.to_datetime(df_wid['date'])

df_wid = df_wid.sort_values(by = 'date') #ordenamos los valores por fecha

df_wid = df_wid.drop(columns=[ 'Impact', 'Affected Population',
                     'Important Person/Group Responsible', 'Month', 'Year', 'Sl. No']) #eliminamos las columnas innecesarias
df_wid.set_index('date', inplace = True) #date sera el index

df_wid
```

El objetivo de haber limpiado ambos datasets es juntarlos por fecha en un solo dataframe final y hacer el análisis de los datos comparando la fecha en la que ocurrió cada evento con el cambio en el precio del petroleo.
